```python
import pandas as pd

planet_df_1 = pd.read_csv('/content/PRO-NASA-Exoplanet-Scraped-Data/updated_scraped_data.csv') #Main page

new_planet_df_1 = pd.read_csv('/content/PRO-NASA-Exoplanet-Scraped-Data/new_scraped_data.csv') #Other info inside specific page

planet_df_1.head(20)
```

```python
new_planet_df_1.head(10)
```

```python
#new_planet_df_1.drop(columns = ['detection_method' , 'mass'] , inplace=True) #Removes columns
new_planet_df_1.head()
```

```python
headers = ["name" , "light_years_from_earth" , "planet_mass" , "stellar_magnitude" , "discovery_date" , "hyperlink" , "planet_type" , "planet_radius" , "orbital_radius" , "orbital_period", "eccentricity" , "detection_method"]

final_planet_df = pd.DataFrame(columns = headers)
final_planet_df = pd.merge(planet_df_1 , new_planet_df_1)
final_planet_df.head()
```
